### What's the Language Model;

Ans：是用来计算一个句子出现的概率的模型。一个语言模型通常构建为字符串s的概率分布p(s)，这里的p(s)实际上反映的是s作为一个句子出现的概率。
这里的概率指的是组成字符串的这个组合，在训练语料中出现的似然，与句子是否合乎语法无关。假设训练语料来自于人类的语言，那么可以认为这个概率是的是一句话是否是人话的概率。

### What's the 1-gram language model

Ans：1-gram语言模型即一元文法模型，它是一种上下文无关模型。表示的是在句子、段落等文本中，该模型仅仅考虑当前词本身出现的概率，而不考虑当前词的上下文环境。概率公式为：$$P(w1,w2,w3...wm) = P(w1)*P(w2)*P(w3)*...*P(wm)$$即一个句子出现的概率等于句子中每个单词概率乘积。

### What's the disadvantages and advantages of 1-gram language model

Ans:优点就是模型简单，计算简便，时间和空间复杂度都相对较低，很快能够得到结果。而由于其考虑的仅仅是词本身出现的概率，但往往在一个句子中，各种词性的词之间都是具有一定的关联性的，比如修饰关系等。所以单纯地考虑本身的概率是一种极其理想化的模型，准确度比较低，得到的结果也比较偏离实际的情况。

### What't the 2-gram models

Ans：即假设文本中的任意一个词出现的概率只同它前面的一个词有关，即某单词出现的概率由其前一个单词来决定。这种假设称为马尔可夫假设。其概率公式为：$$P(S) = P(w1) * P(w2|w1) * P(w3|w2) * ...* P(wn|wn-1)$$

```python
import random
```

```python
def create_grammar(grammar_str , split = "=>" , line_split = "\n"):
    grammar = {}
    for line in grammar_str.split(line_split):
        if not line.strip():
            continue
        exp, stmt = line.split(split)
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar
```

```python
choice = random.choice

def generate(gram,target):
    if target not in gram:
        return target
    expaned = [generate(gram,t) for t in choice(gram[target])]
    return ''.join([e if e!='/n' else '\n' for e in expaned if e!='null'])
```

```python
generate(gram=create_grammar(pets,split='='),target='pets')
```

```python
def generate_n(gram,targ,n,splits="=>"):
    for i in range(n):
        print(generate(gram=create_grammar(gram,split=splits),target=targ))
```

```python
generate_n(pets,"pets",20,splits="=")
```

```python
def prob_1(word):
    return words_count[word] /len(TOKEN)
```

```python
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-1]))]
```

```python
TOKEN_2_GRAM[:10]
```

```python
word_count_2 = Counter(TOKEN_2_GRAM)
```

```python
def prob_2(word1,word2):
    if word1+word2 in word_count_2:
        return word_count_2[word1+word2] / words_count[word2]
    else:
        return 1/len(TOKEN_2_GRAM) #如果不存在组合的词，那么出现频率应该用0表示吗？还是如函数中的else所示，有点疑问
```

```python
prob_2('我们','在')
```

```python
def get_probability(sentence):
    words = cut(sentence)
    sentence_pro = 1
    for i,word in enumerate(words[:-1]):
        next_ = words[i+1]
        probability = prob_2(word,next_)
        sentence_pro *= probability
    #最后需要乘最后一个单词的概率
    pro_last = words_count[words[-1]]/len(TOKEN)
    sentence_pro *= pro_last
    return sentence_pro
```

### 获得最优质的的语言

```python
def get_probability_1(sentence):
    words = cut(sentence)
    sentence_pro = 1
    for i,word in enumerate(words):
        probability = prob_1(word)
        sentence_pro *= probability
    return sentence_pro
```

```python
def get_probability_2(sentence):
    words = cut(sentence)
    sentence_pro = 1
    for i,word in enumerate(words[:-1]):
        next_ = words[i+1]
        probability = prob_2(word,next_)
        sentence_pro *= probability
    #最后需要乘最后一个单词的概率
    pro_last = words_count[words[-1]]/len(TOKEN)
    sentence_pro *= pro_last
    return sentence_pro
```

```python
def generate_best(gram,targets,n,splits="=>",model_gram=2):
    result = []
    for i in range(n):
        if model_gram==1:
            sen = generate(gram=create_grammar(gram,split=splits),target=targets)
            sen = sen.replace('？','')
            prob = get_probability_1(sen)
            result.insert(i,[sen,prob])
        elif model_gram==2:
            sen = generate(gram=create_grammar(gram,split=splits),target=targets)
            sen = sen.replace('？','')
            prob = get_probability(sen)
            result.insert(i,[sen,prob])
    result = sorted(result, key=lambda x: x[1],reverse=True)
    print("The most possible sentence is {} with probability {}.".format(result[0][0],result[0][1]))
    return result
            
        
```

```python
generate_best(relax,"relax",10,splits="=>",model_gram=2)
```

Q: 这个模型有什么问题？ 你准备如何提升？

Ans：
1.计算过程中发现，当出现了别的标点符号的时候，所有的句子的可能性都变成了0，所以在计算的过程中，将问号“？”替换了之后计算的结果，提升的话就是在文本数据清洗的时候，修改正则匹配的范围，将标点符号都囊括进来，这样后面的词组中也会包含这些标点，从而避免了为0的操作；
2.另外是有一些新式的词汇，网络用语无法被囊括，导致计算结果为0，如上图中最后一条的概率就为0，原因是出现了“整吗”，这个现代词汇。所以提升的方式应该是需要增加语料库的储备，多语料库计算来得到更全面的词组关联数据。
3.目前只支持了1-gram和2-gram，可以考虑增添3-gram和4-gram的语言模型，这样的结果可能更为准确。
